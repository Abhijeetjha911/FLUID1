Defining velocity vector

```python
def vel(r,t):
    x=r[0]
    y=r[1]
    x1=x+0.5
    r1= (x1*x1+y*y)**0.5
    if r1< 0.5 :
        u = 0 
        v = 0
    else :
        u = -(x1*x1-y*y)/r1**4
        v = -2*x1*y/r1**4  
    return np.array([u, v])
```
